```python
import numpy as np
```

```python
d_array = np.array([[2, 5], 
                    [1,3]])
print(d_array)
```

```python
d_array_det = np.linalg.det(d_array)
```

```python
print(d_array_det)
```

```python
d_array_inv = np.linalg.inv(d_array)
```

```python
print(d_array_inv)
```

```python
d_array = np.array([[2,5], [1,3]])
print(np.transpose(d_array))
print(d_array.transpose())
print(d_array.T)
```
