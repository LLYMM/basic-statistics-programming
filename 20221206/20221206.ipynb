```python
import numpy as np
A = np.array([[-2,-5,2], [-5,1,-5], [-2,-5,2]])
print("Matrix A: ")
print(A)
print()

eigenvalue, eigenvector = np.linalg.eig(A)
eigenvalue = np.round(eigenvalue, 5)
eigenvector = np.round(eigenvalue , 5)
print("Eigen Value : ")
print(eigenvalue)
print()

print("Eigen Value Diagonal Matrix \u039B(lambda): ")
eigenvalue = np.array([[eigenvalue[0],0,0], [0,eigenvalue[1],0],[0,0,eigenvalue[2]]])
print(eigenvalue)
print()

print("Eigen Vector Matrix P : ")
print(eigenvector)
print()

print("AP = ")
print(np.round(A@eigenvector,0))
print()

print("P\u0389B =")
print(np.round(eigenvector@eigenvalue))
```
