```python
def  gen_permutations(arr, n):
  result = []

  if n == 0:
      return [[]]

  for i, elem in enumerate(arr):
      for P in gen_permutations(arr[:i] + arr[i+1:], n-1):
          result += [[elem]+P]

  return result
```

```python
arr = ['a', 'b', 'c', 'd']

print(gen_permutations(arr, 4))
print(len(gen_permutations(arr, 4)))
```
